```python
def get_details(list_1):
    product_details = {}
    for i in list_1:
        try:
            if i.startswith('made'):
                country = i
                country = country.split(': ')[1]
                product_details['cntry_made_in'] = country
        except:
            pass
    for i in list_1:
        try:
            if i.startswith('collection'):
                col = i
                col = col.split(': ')[1]
                product_details['collection'] = col
        except:
            continue
    for i in list_1:
        try:
            if i.endswith('Rug'):
                size = str(i.split()[:3])
                name = str(i.split()[3:])
                name = name.replace(',','')
                name = name.replace("'","")
                name = name.replace("[",'')
                name = name.replace("]",'')
                product_details['name'] = name
                size = size.replace(',','')
                size = size.replace(' ','')
                size = size.replace('"','')
                size = size.replace("'","")
                size = size.replace("[",'')
                size = size.replace("]",'')
                product_details['size'] = size
        except:
            pass
    for i in list_1:
        try:
            if i.startswith('the primary'):
                color = i
                color = color.split()[4]
                product_details['color'] = color
            else:
                continue
        except:
            pass
    for i in list_1: 
        try:
            if i.startswith('construction'):
                const = i
                const = const.split(': ')[1]
                product_details['construction'] = const
            else:
                continue
        except:
            continue
    for i in list_1:
        try:
            pnt = i.find('%')
            stop = pnt - 2
            price = i[:stop]
            prices = price.replace('$',' ')
            prices = prices.split()
            low_cnt = prices[0] 
            high_cnt = prices[0]
            new_prices = []
            for prc in prices:
                prc = prc.replace(',','')
                prc = int(prc)
                new_prices.append(prc)
                low_cnt = new_prices[0] 
                high_cnt = new_prices[0]
                for prc in new_prices:
                    if prc < low_cnt:
                        low_cnt = prc
                for prc in new_prices:
                    if prc > high_cnt:
                        high_cnt = prc
            sale_price = low_cnt
            orig_price = high_cnt
            product_details['sale_price'] = sale_price
            product_details['orig_price'] = orig_price
        except:
            pass
    for i in list_1:
        try:
            if i.startswith('pile material'):
                material = i
                material = material.split(' ')
                sep = material.index('|')
                start = sep + 3
                ph = material[start:]
                ph = ph[0]
                ph = ph.replace('"','')
                mat = material[2:sep]
                mat = ' '.join(mat)
                product_details['pile_height'] = ph
                product_details['material'] = mat
        except:
            continue
    return product_details
```

```python
reference_dict = {'cntry_made_in':'na','collection':'na','name':'na','size':'na','color':'na',
               'sale_price':'na','orig_price':'na','pile_height':'na','material':'na','construction':'na',}
```

```python
final_product_details = []
new_counter = 0

for url in urls:
    new_counter += 1
    time.sleep(1 + 2*random.random())
    if (new_counter % 1500 == 0):
        time.sleep(320)
    try:
        driver.get(url)
        product_details = []
    except:
        print('no more rugs')
    try:
        elements = driver.find_elements_by_xpath("//*[@id='home']")
        details = [elem.text for elem in elements]
        details_list = details[0].split('\n')
        for i in details_list:
            de = i.lower()
            product_details.append(de)
    except:
        print('could not retrieve details',url)
    try:
        prices_1 = driver.find_element_by_xpath("//*[@id='add_to_cart_form']/div[1]/div/div[2]/div[2]/div").text
        product_details.append(prices_1)
    except:
        print('could not retrieve prices', url)
    try:
        name = driver.find_element_by_xpath("//*[@id='react-root']/div[4]/div[2]/div[2]/div/div[1]/div/h1").text
        product_details.insert(0,name)
    except:
        print('could not retrieve name', url)
    new_dict = get_details(product_details)
    new_dict['url'] = url
    new_dict['product_details'] = product_details
    for k in reference_dict:
        if k not in new_dict:
            new_dict[k] = 'na'
    final_product_details.append(new_dict)
    if (new_counter % 500 == 0):
        with open('working_dictionary.pkl','wb') as f:
            pickle.dump(final_product_details,f)
```
